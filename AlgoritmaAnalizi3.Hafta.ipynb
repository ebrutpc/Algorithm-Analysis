```python
def Maxsubsum(n):
    maxSum = 0 
    for i in range(len(n)):
        for j in range(len(n)):
            thisSum=0
            for k in range(i,j+1):
                thisSum+=n[k]
            if(thisSum > maxSum):
                maxSum = thisSum
    return maxSum
```

```python
def Maxsubsum2(n):
    maxSum=0
    for i in range(len(n)):
        thisSum=0
        for j in range (i,len(n)):
            thisSum+= n[j]
            if(thisSum > maxSum):
                maxSum = thisSum
    return maxSum
```
